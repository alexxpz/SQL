#### Calculate Net Revenue

```python
%%sql

SELECT
  orderdate,
  quantity * netprice / exchangerate as net_revenue
FROM sales
WHERE orderdate::date >= '2020-01-01'
LIMIT 10
```

#### Add Customer & Product Info

```python
%%sql

SELECT
  s.orderdate,
  s.quantity * s.netprice / s.exchangerate as net_revenue,
  c.givenname,
  c.surname,
  c.countryfull,
  c.continent,
  p.productkey,
  p.productname,
  p.categoryname,
  p.subcategoryname,
  CASE WHEN s.quantity * s.netprice / s.exchangerate >= 1000 THEN 'HIGH' ELSE 'LOW' END AS high_low
FROM
  sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
LEFT JOIN product p ON s.productkey = p.productkey
WHERE orderdate::date >= '2020-01-01'
```

#### Basic Aggregation & Pivot

```python
%%sql

SELECT
  orderdate,
  COUNT(DISTINCT(customerkey)) AS total_customers
FROM
  sales
WHERE
  orderdate BETWEEN '2023-01-01' AND '2023-12-31'
GROUP BY
  orderdate
ORDER BY orderdate::DATE
```

```python
%%sql
SELECT DISTINCT
  continent
FROM
  customer
```

```python
%%sql

SELECT
  s.orderdate,
  COUNT(DISTINCT CASE WHEN c.continent = 'Europe' THEN s.customerkey END) AS eu_customers,
  COUNT(DISTINCT CASE WHEN c.continent = 'North America' THEN s.customerkey END) AS na_customers,
  COUNT(DISTINCT CASE WHEN c.continent = 'Australia' THEN s.customerkey END) AS au_customers
FROM
  sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
WHERE
  s.orderdate BETWEEN '2023-01-01' AND '2023-12-31'
GROUP BY
  s.orderdate
ORDER BY
  s.orderdate::DATE
```

```python
%%sql

SELECT
  p.categoryname,
  SUM(CASE WHEN s.orderdate BETWEEN '2022-01-01' AND '2022-12-31' THEN s.quantity * s.netprice / s.exchangerate END) AS "total_net_revenue_2022",
  SUM(CASE WHEN s.orderdate BETWEEN '2023-01-01' AND '2023-12-31' THEN s.quantity * s.netprice / s.exchangerate END) AS "total_net_revenue_2023"
FROM
  sales s
LEFT JOIN product p ON s.productkey = p.productkey
GROUP BY
  p.categoryname
```

#### Statistical analysis

```python
%%sql

SELECT
  p.categoryname,
  AVG(CASE WHEN s.orderdate BETWEEN '2022-01-01' AND '2022-12-31' THEN s.quantity * s.netprice / s.exchangerate END) AS "average_order_net_revenue_2022",
  AVG(CASE WHEN s.orderdate BETWEEN '2023-01-01' AND '2023-12-31' THEN s.quantity * s.netprice / s.exchangerate END) AS "average_order_net_revenue_2023"
FROM
  sales s
LEFT JOIN product p ON s.productkey = p.productkey
GROUP BY
  p.categoryname
```

```python
%%sql

SELECT
  p.categoryname AS category,
  PERCENTILE_CONT(0.50) WITHIN GROUP(ORDER BY (CASE WHEN s.orderdate BETWEEN '2022-01-01' AND '2022-12-31' THEN s.quantity * s.netprice / s.exchangerate END)) AS y2022_median_sales,
  PERCENTILE_CONT(0.50) WITHIN GROUP(ORDER BY (CASE WHEN s.orderdate BETWEEN '2023-01-01' AND '2023-12-31' THEN s.quantity * s.netprice / s.exchangerate END)) AS y2023_median_sales
FROM
  sales s
LEFT JOIN product p ON s.productkey = p.productkey
GROUP BY
  p.categoryname
```

#### Advanced Segmentation

##### Categorize

```python
%%sql

SELECT
  orderdate,
  quantity,
  netprice,
  CASE WHEN quantity >=2 AND netprice >= 100 THEN 'Multiple High Value Items'
       WHEN quantity >=2 THEN 'Multiple Standard Items'
       WHEN netprice >= 100 THEN 'Single High Value Item'
       ELSE 'Single Standard Item'
  END AS order_type
FROM sales
LIMIT 20
```

```python
%%sql

WITH median_value AS (
  SELECT
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY s.quantity * s.netprice / s.exchangerate)
  FROM
    sales s
  WHERE
    s.orderdate BETWEEN '2022-01-01' AND '2023-12-31'
)

SELECT
  p.categoryname,
  SUM(CASE WHEN ((s.quantity * s.netprice / s.exchangerate) < mv AND s.orderdate BETWEEN '2022-01-01' AND '2022-12-31') THEN s.quantity * s.netprice / s.exchangerate END) AS y2022_low_value_orders_net_revenue,
  SUM(CASE WHEN ((s.quantity * s.netprice / s.exchangerate) >= mv AND s.orderdate BETWEEN '2022-01-01' AND '2022-12-31') THEN s.quantity * s.netprice / s.exchangerate END) AS y2022_high_value_orders_net_revenue,
  SUM(CASE WHEN ((s.quantity * s.netprice / s.exchangerate) < mv AND s.orderdate BETWEEN '2023-01-01' AND '2023-12-31') THEN s.quantity * s.netprice / s.exchangerate END) AS y2023_low_value_orders_net_revenue,
  SUM(CASE WHEN ((s.quantity * s.netprice / s.exchangerate) >= mv AND s.orderdate BETWEEN '2023-01-01' AND '2023-12-31') THEN s.quantity * s.netprice / s.exchangerate END) AS y2023_high_value_orders_net_revenue
FROM
  sales s
  LEFT JOIN product p ON s.productkey = p.productkey
  median_value mv
GROUP BY
  p.categoryname
```
